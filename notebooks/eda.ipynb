```python
processed_data_path = "../data/fake_news/processed/gossipcop_processed.csv"
```

```python
import pandas as pd

# Load the processed data
df = pd.read_csv(processed_data_path)   

# Display the first few rows of the dataframe
df.head()
```

```python
# get the avarage length of the text
average_length = df['title'].apply(lambda x: len(x.split())).mean()
print(f"Average length of text: {average_length} words")
```

```python
# plot the distribution of the text length
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['title'].apply(lambda x: len(x.split())), bins=30, kde=True)
plt.title('Distribution of Text Length')
plt.xlabel('Length of Text (in words)')
plt.ylabel('Frequency')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def plot_word_cloud(text, title):
    """Generate and plot a word cloud."""
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(text))
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title, fontsize=16)
    plt.show()

# Filter data for label 0 (real news) and label 1 (fake news)
real_news = df[df['label'] == 0]['title']
fake_news = df[df['label'] == 1]['title']

# Plot word clouds
plot_word_cloud(real_news, "Word Cloud for Real News (Label 0)")
plot_word_cloud(fake_news, "Word Cloud for Fake News (Label 1)")
```
